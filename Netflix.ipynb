```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 🔧 Step 1: Create mock Netflix-style data manually
data = {
    'title': [
        'Stranger Things', 'The Irishman', 'Dark', 'Extraction', 'BoJack Horseman',
        'Lupin', 'The Witcher', '13 Reasons Why', 'Narcos', 'Money Heist'
    ],
    'type': ['TV Show', 'Movie', 'TV Show', 'Movie', 'TV Show', 
             'TV Show', 'TV Show', 'TV Show', 'TV Show', 'TV Show'],
    'country': ['United States', 'United States', 'Germany', 'India', 'United States',
                'France', 'Poland', 'United States', 'Colombia', 'Spain'],
    'release_year': [2016, 2019, 2017, 2020, 2014, 
                     2021, 2019, 2017, 2015, 2017],
    'listed_in': [
        'Drama, Sci-Fi', 'Crime, Drama', 'Thriller, Sci-Fi', 'Action, Thriller', 'Comedy, Animation',
        'Mystery, Crime', 'Fantasy, Drama', 'Teen, Drama', 'Crime, Thriller', 'Action, Crime'
    ],
    'duration': [
        '3 Seasons', '209 min', '3 Seasons', '117 min', '6 Seasons', 
        '2 Seasons', '2 Seasons', '4 Seasons', '3 Seasons', '5 Seasons'
    ]
}

# Convert to DataFrame
df = pd.DataFrame(data)

# ✅ Step 2: Basic overview
print("📄 Data Preview:")
print(df)

# ✅ Step 3: Content type distribution
sns.countplot(data=df, x='type', palette='Set2')
plt.title("Distribution of Movies vs TV Shows")
plt.xlabel("Type")
plt.ylabel("Count")
plt.show()

# ✅ Step 4: Content released per year
plt.figure(figsize=(10,5))
sns.histplot(data=df, x='release_year', bins=10, color='skyblue')
plt.title("Content Released by Year")
plt.xlabel("Release Year")
plt.ylabel("Number of Titles")
plt.show()

# ✅ Step 5: Top countries (count manually for now)
country_counts = df['country'].value_counts()
print("\n🌍 Content count by country:\n", country_counts)

country_counts.plot(kind='bar', color='orange')
plt.title("Top Countries by Number of Titles")
plt.xlabel("Country")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# ✅ Step 6: Genre analysis
genre_series = df['listed_in'].str.split(', ')
all_genres = genre_series.explode()
genre_counts = all_genres.value_counts()
print("\n🎭 Genre count:\n", genre_counts)

genre_counts.plot(kind='bar', color='green')
plt.title("Most Common Genres")
plt.xlabel("Genre")
plt.ylabel("Frequency")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# ✅ Step 7: Movie duration analysis (only for 'Movie' type)
movies = df[df['type'] == 'Movie'].copy()
movies['duration_mins'] = movies['duration'].str.extract('(\d+)').astype(float)

plt.figure(figsize=(8,5))
sns.histplot(movies['duration_mins'], bins=5, color='purple')
plt.title("Movie Duration Distribution")
plt.xlabel("Duration (minutes)")
plt.ylabel("Number of Movies")
plt.show()
```
